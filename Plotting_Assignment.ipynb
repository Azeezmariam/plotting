```python
# YOUR CODE HERE
# Plot a histogram

# Set the figure size
plt.figure(figsize=(8, 6))

# Plot the histogram with different bin widths
plt.hist(data['co2'], bins=30, color='grey', alpha=0.7,edgecolor="black")

# Add labels and title
plt.title('Distribution of carbon Dioxide of vehicles')
plt.xlabel('Carbon Dioxide  (g/mi)')
#g/mi" stands for "grams per mile." It represents the amount of carbon dioxide (CO2) emissions produced per mile traveled by a vehicle
plt.ylabel('Frequency')
plt.savefig('co2_vehicles')
plt.show()


```

```python
# YOUR CODE HERE
# Plot a histogram

# Set the figure size
plt.figure(figsize=(8, 6))

# Plot the histogram with different bin widths
plt.hist(data['comb'], bins=20, color='yellow', alpha=0.7, edgecolor='black')

# Add labels and title
plt.title('Histogram of Fuel Efficiency')
plt.xlabel('Combined Fuel Efficiency (mpg)')
plt.ylabel('Frequency')
#mpg is miles per galon
plt.savefig('comb_fuel_efficiency')

# Show the plot
plt.show()
```

```python
# Select only numeric columns
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_data.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.xticks(rotation=45)
plt.title('Correlation Heatmap')
plt.savefig('heatmap')
plt.show()
```

```python
!mv heatmap.png chart_images
```

INTERPRETATION:
The code selects numeric data from the dataset, including both float and integer types. It then computes the correlation matrix among these numerical variables. The heatmap visually represents this correlation matrix, with each cell color-coded based on the correlation coefficient between the variables it corresponds to. Annotations are displayed within each cell, indicating the correlation coefficient values. The colormap 'coolwarm' is used to highlight positive and negative correlations effectively. Additionally, the fmt=".2f" ensures that correlation coefficients are displayed as floating-point numbers with two decimal places. Overall, this heatmap provides insights into the relationships between different numeric variables, aiding in understanding their interdependencies and potential predictive power.
